```python
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('Online-eCommerce.csv')
data = data.dropna()
```

```python
display(data.head())
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

numeric_features = data[['Cost', 'Sales', 'Quantity', 'Total_Cost', 'Total_Sales']].dropna()
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_features)

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(scaled_data)


def plot_clusters(df, labels, title):
    plt.figure(figsize=(8, 6))
    plt.scatter(df[:, 0], df[:, 1], c=labels, cmap='viridis', s=50)
    plt.title(title)
    plt.show()

plot_clusters(scaled_data, kmeans_labels, 'K-means Clustering')
```

```python
from sklearn.model_selection import train_test_split

features = ['Cost', 'Sales', 'Quantity', 'Total_Cost', 'Total_Sales']
X = data[features]
y = data['Category']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

print(f"Тренувальний набір: X_train: {X_train.shape}, y_train: {y_train.shape}")
print(f"Тестовий набір: X_test: {X_test.shape}, y_test: {y_test.shape}")
```

```python
fig, axes = plt.subplots(1, 1, figsize=(8, 6))

axes.scatter(X_train['Cost'], X_train['Sales'], 
             c='blue', label="Тренувальний набір", s=60)

axes.scatter(X_test['Cost'], X_test['Sales'], 
             marker='^', c='red', label="Тестовий набір", s=60)

axes.legend(loc='upper left')
axes.set_title("Тренувальний та тестовий набори (до масштабування)")
axes.set_xlabel("Cost")
axes.set_ylabel("Sales")

plt.show()

# лінійна залежність
```

```python
# Побудова графіка з іншими ознаками
fig, axes = plt.subplots(1, 1, figsize=(8, 6))
axes.scatter(X_train['Quantity'], X_train['Total_Sales'], 
             c='blue', label="Тренувальний набір", s=60)
axes.scatter(X_test['Quantity'], X_test['Total_Sales'], 
             marker='^', c='red', label="Тестовий набір", s=60)
axes.legend(loc='upper left')
axes.set_title("Тренувальний та тестовий набори (інші ознаки)")
axes.set_xlabel("Quantity")
axes.set_ylabel("Total_Sales")
plt.show()

# висока лінійність
```

```python
import numpy as np

category_ssd = data[data['Category'] == 'SSD']
category_cpu = data[data['Category'] == 'CPU']

# Вибрані ознаки для побудови гістограм
features = ['Cost', 'Sales', 'Quantity', 'Total_Cost', 'Total_Sales']

# Створюємо підграфіки для кожної ознаки
fig, axes = plt.subplots(3, 2, figsize=(12, 10))
ax = axes.ravel()

# Будуємо гістограми для кожної ознаки
for i, feature in enumerate(features):
    _, bins = np.histogram(data[feature], bins=50)  # Розрахунок бінів
    ax[i].hist(category_ssd[feature], bins=bins, color='blue', alpha=0.5, label='SSD')
    ax[i].hist(category_cpu[feature], bins=bins, color='red', alpha=0.5, label='CPU')
    ax[i].set_title(feature)
    ax[i].set_yticks(())

# Додаємо підписи та легенду
ax[0].set_xlabel("Значення ознаки")
ax[0].set_ylabel("Частота")
ax[0].legend(loc="best")
fig.tight_layout()

# Показуємо графік
plt.show()
```

```python
X = data[['Cost', 'Sales', 'Quantity']].dropna()

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=2, random_state=1)
kmeans.fit(X_scaled)

data['Cluster'] = kmeans.labels_

sse = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=1)
    kmeans.fit(X_scaled)
    sse.append(kmeans.inertia_)

plt.plot(range(1, 10), sse, marker='o')
plt.xlabel("Кількість кластерів")
plt.ylabel("SSE (сума квадратів відстаней)")
plt.title("Метод ліктя")
plt.show()
```

```python
from sklearn.decomposition import PCA

# Застосування PCA для зменшення до 2 вимірів
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Кластеризація на основі зменшених вимірів
kmeans_pca = KMeans(n_clusters=3, random_state=1)
kmeans_pca.fit(X_pca)

# Візуалізація результатів
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans_pca.labels_, cmap='viridis', alpha=0.6)
plt.title("Кластеризація після PCA")
plt.xlabel("Перша головна компонента")
plt.ylabel("Друга головна компонента")
plt.colorbar(label='Кластер')
plt.show()
```

```python
# Перетворюємо колонку Order_Date у формат datetime
data['Order_Date'] = pd.to_datetime(data['Order_Date'], errors='coerce')

# Перевіряємо, чи немає пропущених значень після перетворення
print(data['Order_Date'].isna().sum())  # Якщо є NaN, це можуть бути некоректні дати
data = data.dropna(subset=['Order_Date'])

# Групуємо дані за іменами клієнтів та обчислюємо ключові показники
customer_data = data.groupby('Customer_Name').agg({
    'Total_Sales': 'sum',         # Загальна сума продажів
    'Quantity': 'sum',            # Загальна кількість проданих товарів
    'Cost': 'mean',               # Середня вартість товару
    'Category': lambda x: x.mode()[0],  # Найчастіша категорія товарів
    'Order_Date': lambda x: (pd.Timestamp.now() - x.max()).days  # Дні з останньої покупки
}).reset_index()

# Перейменовуємо колонки для зручності
customer_data.columns = [
    'Customer_Name', 'Total_Spent', 'Total_Quantity', 'Avg_Cost', 
    'Favorite_Category', 'Days_Since_Last_Purchase'
]

display(customer_data.head())
```

```python
from sklearn.preprocessing import StandardScaler

# Вибираємо числові ознаки для кластеризації
X = customer_data[['Total_Spent', 'Total_Quantity', 'Avg_Cost', 'Days_Since_Last_Purchase']]

# Масштабуємо дані
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans

sse = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=1)
    kmeans.fit(X_scaled)
    sse.append(kmeans.inertia_)

plt.plot(range(1, 10), sse, marker='o')
plt.xlabel("Кількість кластерів")
plt.ylabel("SSE (сума квадратів відстаней)")
plt.title("Метод ліктя")
plt.show()
```

```python
# Ініціалізуємо модель K-means з оптимальною кількістю кластерів
kmeans = KMeans(n_clusters=2, random_state=1)
kmeans.fit(X_scaled)

# Додаємо кластери в таблицю з даними
customer_data['Cluster'] = kmeans.labels_

# Перевіряємо результат
print(customer_data[['Customer_Name', 'Cluster']].head())
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Зменшуємо розмірність до 2 компонентів
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Візуалізуємо результати кластеризації
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=customer_data['Cluster'], cmap='viridis', alpha=0.6)
plt.title("Кластеризація покупців на основі вподобань")
plt.xlabel("Перша головна компонента")
plt.ylabel("Друга головна компонента")
plt.colorbar(label='Кластер')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# Обчислюємо коефіцієнт силуету
silhouette_avg = silhouette_score(X_pca, kmeans.labels_)
print(f"Середній коефіцієнт силуету: {silhouette_avg:.2f}")
```
