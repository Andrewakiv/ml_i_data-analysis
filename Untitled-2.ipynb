```python
total_sales = data['Total_Sales'].dropna()

Q1 = total_sales.quantile(0.25)
Q3 = total_sales.quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

total_sales_filtered = total_sales[(total_sales >= lower) & (total_sales <= upper)]

total_sales_filtered_mean = total_sales_filtered.mean()
total_sales_filtered_median = total_sales_filtered.median()
total_sales_filtered_mode = total_sales_filtered.mode()[0]
total_sales_filtered_range = total_sales_filtered.max() - total_sales.min()
total_sales_filtered_quantiles = total_sales_filtered.quantile([0.25, 0.5, 0.75])
total_sales_filtered_variance = total_sales_filtered.var()

{
    'середнє': total_sales_filtered_mean,
    'медіана': total_sales_filtered_median,
    'мода (режим)': total_sales_filtered_mode,
    'діапазон': total_sales_filtered_range,
    'квартилі': total_sales_filtered_quantiles,
    'дисперсія': total_sales_filtered_variance
}
```

```python
missing_values = data['Total_Sales'].isnull().sum()
print(f"Пропущені значення: {missing_values}")

invalid_values = data[data['Total_Sales'] <= 0]
print(f"Кількість нульових або негативних значень: {len(invalid_values)}")

Q1 = data['Total_Sales'].quantile(0.25)
Q3 = data['Total_Sales'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = data[(data['Total_Sales'] < lower) | (data['Total_Sales'] > upper)]
print(f"Кількість викидів: {len(outliers)}")
```

```python
incorrec_sales_data_consistency = data[data['Total_Sales'] != data['Sales'] * data['Quantity']]
print(f'incorrect data count: {len(incorrec_sales_data_consistency)}')


total_sales_without_null = data['Total_Sales'] != None
sales_without_null = data['Sales'] != None
quantity_without_null = data['Quantity'] != None
sales_data_consistency = data[total_sales_without_null != sales_without_null * quantity_without_null]
print(f'incorrect data count: {len(sales_data_consistency)}')
```

```python
total_sales = data['Total_Sales'].dropna()

quantile_bins = pd.qcut(total_sales, q=4)

equal_width_bins = pd.cut(total_sales, bins=4)

binned_data = total_sales.groupby(equal_width_bins).mean()

min_max_normalized = (total_sales - total_sales.min()) / (total_sales.max() - total_sales.min())

z_score_normalized = (total_sales - total_sales.mean()) / total_sales.std()

{
    "Розбиття на рівні частоти (квартилі)": display(quantile_bins.head()),
    "Розбиття на рівні інтервали": display(equal_width_bins.head()),
    "Згладжені дані (середнє в кожному сегменті)": display(binned_data.head()),
    "Мінімакс нормалізація": display(min_max_normalized.head()),
    "Z-score нормалізація": display(z_score_normalized.head())
}
```

```python
data_cleaned = data.copy()
numeric_cols = data.select_dtypes(include=['float64', 'int64']).columns
data_cleaned[numeric_cols] = data_cleaned[numeric_cols].fillna(data_cleaned[numeric_cols].mean())

scaler = MinMaxScaler()
data_cleaned['Quantity_MinMax'] = scaler.fit_transform(data_cleaned[['Quantity']])

scaler_z = StandardScaler()
data_cleaned['Sales_Zscore'] = scaler_z.fit_transform(data_cleaned[['Sales']])

display(data_cleaned[['Quantity', 'Quantity_MinMax', 'Sales', 'Sales_Zscore']].head())
```
