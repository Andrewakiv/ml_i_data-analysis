```python
import pandas as pd
import matplotlib.pyplot as plt

# Дані для K-Means
kmeans_data = {
    "Cluster": [0, 1, 2],
    "Element Count": [5568, 4742, 4690],
    "Average Age": [43.289, 43.404, 43.515],
    "Purchase Amount": [255.441, 251.6797, 252.1333]
}

# Дані для EM
em_data = {
    "Cluster": [0, 1, 2, 3],
    "Element Count": [1654, 4690, 4636, 4020],
    "Average Age": [43.3311, 43.5266, 43.4634, 43.226],
    "Purchase Amount": [249.2563, 252.2378, 251.9558, 256.7522]
}

# Завантаження даних у DataFrames
df_kmeans = pd.DataFrame(kmeans_data)
df_em = pd.DataFrame(em_data)

# Перший графік: Гістограма кількості елементів у кожному кластері для K-Means
plt.figure(figsize=(10, 6))
plt.bar(df_kmeans["Cluster"], df_kmeans["Element Count"], color='skyblue')
plt.xlabel("Cluster")
plt.ylabel("Element Count")
plt.title("K-Means: Element Count per Cluster")
plt.show()

# Другий графік: Гістограма кількості елементів у кожному кластері для EM
plt.figure(figsize=(10, 6))
plt.bar(df_em["Cluster"], df_em["Element Count"], color='lightgreen')
plt.xlabel("Cluster")
plt.ylabel("Element Count")
plt.title("EM: Element Count per Cluster")
plt.show()

# Третій графік: Лінійна діаграма середнього віку у кожному кластері для K-Means
plt.figure(figsize=(10, 6))
plt.plot(df_kmeans["Cluster"], df_kmeans["Average Age"], marker='o', color='blue')
plt.xlabel("Cluster")
plt.ylabel("Average Age")
plt.title("K-Means: Average Age per Cluster")
plt.show()

# Четвертий графік: Лінійна діаграма середнього віку у кожному кластері для EM
plt.figure(figsize=(10, 6))
plt.plot(df_em["Cluster"], df_em["Average Age"], marker='o', color='green')
plt.xlabel("Cluster")
plt.ylabel("Average Age")
plt.title("EM: Average Age per Cluster")
plt.show()

# П'ятий графік: Лінійна діаграма середньої суми покупок у кожному кластері для K-Means
plt.figure(figsize=(10, 6))
plt.plot(df_kmeans["Cluster"], df_kmeans["Purchase Amount"], marker='o', color='blue')
plt.xlabel("Cluster")
plt.ylabel("Purchase Amount")
plt.title("K-Means: Average Purchase Amount per Cluster")
plt.show()

# Шостий графік: Лінійна діаграма середньої суми покупок у кожному кластері для EM
plt.figure(figsize=(10, 6))
plt.plot(df_em["Cluster"], df_em["Purchase Amount"], marker='o', color='green')
plt.xlabel("Cluster")
plt.ylabel("Purchase Amount")
plt.title("EM: Average Purchase Amount per Cluster")
plt.show()
```
