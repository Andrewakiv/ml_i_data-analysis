```python
# Завантажуємо дані
data = pd.read_csv('Online-eCommerce.csv')

# Вибираємо необхідні колонки (наприклад, Customer_Name і Category)
transactions = data.groupby(['Customer_Name', 'Category'])['Category'].count().unstack().reset_index().fillna(0)

# Перетворюємо кількості на бінарний формат (1 - товар куплено, 0 - не куплено)
transactions = transactions.set_index('Customer_Name')
transactions = transactions.applymap(lambda x: 1 if x > 0 else 0)

display(transactions.head())  # Перевіряємо, чи дані підготовлені
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

# Використовуємо Apriori для знаходження частих наборів товарів
frequent_itemsets = apriori(transactions, min_support=0.05, use_colnames=True)

# Генеруємо асоціативні правила на основі частих наборів
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Виводимо результати
display(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

# Використовуємо Apriori для знаходження частих наборів товарів
frequent_itemsets = apriori(transactions, min_support=0.05, use_colnames=True)

# Генеруємо асоціативні правила на основі частих наборів
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Відбираємо топ-10 правил за значенням lift
top_lift_rules = rules.sort_values(by='lift', ascending=False).head(10)

# Відбираємо топ-10 правил за значенням confidence
top_confidence_rules = rules.sort_values(by='confidence', ascending=False).head(10)

# Виводимо результати для топ-правил
print("Топ-10 правил за Lift:")
display(top_lift_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])

print("\nТоп-10 правил за Confidence:")
display(top_confidence_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```

```python
# Завантаження даних та підготовка транзакцій із групуванням за користувачем та датою
import pandas as pd

# Завантаження даних
data = pd.read_csv('Online-eCommerce.csv')

# Перетворення 'Order_Date' у формат datetime
data['Order_Date'] = pd.to_datetime(data['Order_Date'], errors='coerce')
data = data.dropna(subset=['Order_Date'])  # Видалення некоректних дат

# Групування за користувачем та датою покупки з категоріями
transactions = data.groupby(['Customer_Name', data['Order_Date'].dt.date, 'Category']).size().unstack(fill_value=0)

# Перетворення кількостей на бінарний формат (1 - товар куплено, 0 - не куплено)
transactions = transactions.applymap(lambda x: 1 if x > 0 else 0)

transactions.head(100)  # Перевіряємо результат
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

# Використання алгоритму Apriori для знаходження частих наборів товарів
frequent_itemsets = apriori(transactions, min_support=0.01, use_colnames=True)

# Генерація асоціативних правил на основі частих наборів
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Виведення правил
display(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```
