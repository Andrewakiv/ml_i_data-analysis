```python
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

data = pd.read_csv('Online-eCommerce.csv')
data = data.dropna()
```

```python
data['Order_Date'] = pd.to_datetime(data['Order_Date'], errors='coerce')

print(data['Order_Date'].isna().sum())
data = data.dropna(subset=['Order_Date'])

customer_data = data.groupby('Customer_Name').agg({
    'Total_Sales': 'sum',         
    'Quantity': 'sum',            
    'Cost': 'mean',               
    'Category': lambda x: x.mode()[0],  
    'Order_Date': lambda x: (pd.Timestamp.now() - x.max()).days
}).reset_index()


customer_data.columns = [
    'Customer_Name', 'Total_Spent', 'Total_Quantity', 'Avg_Cost', 
    'Favorite_Category', 'Days_Since_Last_Purchase'
]

display(customer_data.head())
```

```python
from sklearn.preprocessing import StandardScaler

X = customer_data[['Total_Spent', 'Total_Quantity', 'Avg_Cost', 'Days_Since_Last_Purchase']]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
sse = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=1)
    kmeans.fit(X_scaled)
    sse.append(kmeans.inertia_)

plt.plot(range(1, 10), sse, marker='o')
plt.xlabel("Кількість кластерів")
plt.ylabel("SSE (сума квадратів відстаней)")
plt.title("Метод ліктя")
plt.show()
```

```python

kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X_scaled)


customer_data['Cluster'] = kmeans.labels_

plt.figure(figsize=(8, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=customer_data['Cluster'], cmap='viridis', alpha=0.6)
plt.title("Кластеризація k-mean")
plt.xlabel("Total_Spent (масштабований)")
plt.ylabel("Total_Quantity (масштабований)")
plt.colorbar(label='Кластер')
plt.show()
```

```python
display(customer_data.head(100))
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X_pca)

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=customer_data['Cluster'], cmap='viridis', alpha=0.6)
plt.title("Кластеризація покупців з PCA")
plt.xlabel("Перша головна компонента")
plt.ylabel("Друга головна компонента")
plt.colorbar(label='Кластер')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

silhouette_avg = silhouette_score(X_scaled, kmeans.labels_)
print(f"Середній коефіцієнт силуету: {silhouette_avg:.2f}")
silhouette_avg = silhouette_score(X_pca, kmeans.labels_)
print(f"Середній коефіцієнт силуету: {silhouette_avg:.2f}")
```
